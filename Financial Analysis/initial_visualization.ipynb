## Shifting to calculate change

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
# shifting to one day in order to see the change
aapl["lag1"] = aapl.shift(periods = 1)
```

```python
aapl.head()
```

```python
# subtract for change
aapl.AAPL.sub(aapl.lag1)
```

```python
# creating a column for the daily difference
aapl["daily_diff"]= aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
#percent change
aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl["pct_change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

## Calculating absolute and relative change


```python
aapl.head()
```

```python
aapl.AAPL.diff(periods = 1)
```

```python
aapl["abs_diff"]= aapl.AAPL.diff(periods=1)
```

```python
aapl.head()
```

```python
aapl.daily_diff.equals(aapl.abs_diff)
```

```python
aapl["rel_diff"] = aapl.AAPL.pct_change(periods = 1).mul(100)
```

```python
aapl.head()
```

```python
#BM last business day of the month
monthly_returns = aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
monthly_returns.plot(kind = "line", figsize = (20, 8),fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

# Stock performance measured using Mean return and STD of returns

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
ret = aapl.pct_change().dropna()
```

```python
ret.head()
```

```python
ret.info()
```

```python
ret.plot(kind = "hist", figsize = (12,8), bins = 100)
```

```python
daily_mean_return = ret.mean()
daily_mean_return
```

```python
var_daily_ret = ret.var()
var_daily_ret
```

```python
std_daily_ret = np.sqrt(var_daily_ret)
std_daily_ret
```

```python
# Standard dev returns
ret.std()
```

```python
# annual mean return pct
annual_mean_ret = ret.mean()*252
annual_mean_ret
```

```python
# annual_variation return
annual_var_ret = ret.var()*252
annual_var_ret
```

```python
# annual std of returns return
annual_std_returns = np.sqrt(annual_var_ret)
annual_std_returns
```

```python
# annual std return
ret.std()*np.sqrt(252)
```

# Financial Time Series Return vs Risk

```python
norm_pricing.plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show
```

```python
close.head()
```

```python
close.pct_change()
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
ret.describe()
```

```python
# transposing the describe df to isolate the mean and the standard dev
ret.describe().T.loc[:, ["mean", "std"]]
```

```python
summary = ret.describe().T.loc[:, ["mean", "std"]]
```

```python
# annualized returns vs risk
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
```

```python
summary.head()
```

```python
summary.plot.scatter(x = "std", y = "mean", figsize = (12,8), s=50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy= (summary.loc[i,"std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("annualized Risk(std)", fontsize = 15)
plt.ylabel("annualized Return(std)", fontsize = 15)
plt.title("Risk/Return",fontsize = 20)
plt.show
```
