## Initial visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# fixing the index back to date
stocks = pd.read_csv("stocks.csv", header= [0,1], index_col = [0], parse_dates = [0])
```

```python
close= stocks.loc[:, "Close"].copy()
```

```python
# visualizing charts initially
close.plot(figsize = (20, 8),fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

## Normalizing Time series to a base value of 100

```python
close.head()
```

```python
# visializing first stock first column
close.iloc[0,0]
```

```python
# adjusting and normalizing starting point for apple 
close.AAPL.div(close.iloc[0,0]).mul(100)
```

```python
# normalizing all the first close prices
norm_pricing = close.div(close.iloc[0]).mul(100)
```

```python
# plotting normalized
norm_pricing.plot(figsize = (20, 8),fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```
