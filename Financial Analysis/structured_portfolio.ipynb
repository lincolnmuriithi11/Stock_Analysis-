## Shifting to calculate change

```python
vig = close.VIG.copy().to_frame()
```

```python
vig.head()
```

```python
# shifting to one day in order to see the change
vig["lag1"] = vig.shift(periods = 1)
```

```python
vig.head()
```

```python
# subtract for change
vig.VIG.sub(vig.lag1)
```

```python
# creating a column for the daily difference
vig["daily_diff"]= vig.VIG.sub(vig.lag1)
```

```python
vig.head()
```

```python
#percent change
vig.VIG.div(vig.lag1).sub(1).mul(100)
```

```python
vig["pct_change"] = vig.VIG.div(vig.lag1).sub(1).mul(100)
```

```python
vig.head()
```

```python
ret.head()
```

```python
ret.describe()
```

```python
# transposing the describe df to isolate the mean and the standard dev
ret.describe().T.loc[:, ["mean", "std"]]
```

```python
summary = ret.describe().T.loc[:, ["mean", "std"]]
```

```python
# annualized returns vs risk
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
```

```python
summary.head()
```

```python
summary.plot.scatter(x = "std", y = "mean", figsize = (12,8), s=50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy= (summary.loc[i,"std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("annualized Risk(std)", fontsize = 15)
plt.ylabel("annualized Return", fontsize = 15)
plt.title("Risk/Return",fontsize = 20)
plt.show
```
