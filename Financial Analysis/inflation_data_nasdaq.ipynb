```python
import pandas as pd
tables = [f"RATEINF/{region}" for region in pd.DataFrame(ds.to_list())[1].to_list()]
```

```python
tables.sort()
```

```python
tables
```

```python
data_cpi = []
data_inflation = []
for region in tables:
  df_tmp = nasdaqdatalink.get(region, paginate=True)
  df_tmp.columns = [region.split('/')[-1]]
  if 'CPI' in region:
    data_cpi.append(df_tmp)
  else:
    data_inflation.append(df_tmp)
```

```python
pd.concat(data_cpi).sort_index().ffill().plot(figsize=(18, 9))
```

```python
pd.concat(data_inflation).sort_index().ffill().plot(figsize=(18, 40), subplots=True)
```
